```python
in_file="/home/xuefei/CSCI5502-DataMining_Project/Dataset/employment/Contributing_family _workers_female.xlsx"
oldwb = xl.load_workbook(in_file)
oldws = oldwb.worksheets[0]
```

```python
in_file1="/home/xuefei/CSCI5502-DataMining_Project/Dataset/employment/real_data.xlsx"
oldwb1 = xl.load_workbook(in_file1)
oldws1 = oldwb1.worksheets[0]
```

```python
female_family_worker={}
male_family_worker={}
total_count=0
total_count1=0
for process_line in range(2,189):
    count=0
    total=0
    count1=0
    total1=0
    country_name=oldws.cell(row = process_line, column = 1).value
    real_index=None
    for row_index in range(2,265):
        if oldws1.cell(row= row_index, column = 1).value==country_name:
            real_index=row_index
            print(real_index)
            
    for cols in range(3,33):
        if oldws.cell(row = process_line, column = cols).value!=None:
            count=count+1
            total=total+oldws.cell(row = process_line, column = cols).value
        if oldws1.cell(row = real_index, column = cols).value!=None:
            count1=count1+1
            total1=total1+oldws1.cell(row = real_index, column = cols).value
    mean=total/count
    mean1=total1/count1
    total_count=total_count+total
    total_count1=total_count1+total1
    print(count)     
    print(total)
    print(mean)
    print(count1)     
    print(total1)
    print(mean1)
    print(country_name)
    female_family_worker[country_name]=mean
    male_family_worker[country_name]=mean1
```
